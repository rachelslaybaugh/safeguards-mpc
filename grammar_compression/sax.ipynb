# Symbolic Aggregate approXimation (SAX)

With the fundamentals of time series covered in [the first notebook](time-series.ipynb), we can move on to the process of representing them as symbols, which will ultimately allow them to be treated as a grammar and compressed. The process is fully described in the paper by [Lin, et al. (Jun. 2003)](http://www.cs.ucr.edu/~eamonn/SAX.pdf). The code in this notebook relies on several modules in the `time_series_utils` package, which has classes and functions for manipulating time series.

```python
# Overhead
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
from time_series_utils.comparisons import *
from time_series_utils.vectorize import *
from time_series_utils.time_series import TimeSeries,Subsequence
```

We will begin with a reminder of the definition of a time series. A time series $T$ of length $m$ is given by 

$$ T = T_1, ..., T_m $$

where $T_i$ is the $i^{\text{th}}$ measurement of $T$. 

We will use a similar time series to the one defined in the time series notebook ,but with many less datapoints.

```python
# Create a time series with m observations
m = 200
times = np.arange(m)+1
periods = 5
multiplier = 0.5+0.5*signal.sawtooth(np.linspace(0,np.pi*periods,m),0)
observations = 100 + (20*np.sin(times*2*np.pi*periods/m)*multiplier) + 2*np.random.normal(0,1,m)
T = TimeSeries(times,observations)
# Plot the time series
T.plot_timeseries(f'Time Series $T$, with $m={m}$ observations')
plt.show()
```
